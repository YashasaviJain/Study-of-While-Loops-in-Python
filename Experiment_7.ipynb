```python
# Wap to print i as long as i is less than 6
i=1
while i<=6:
  print(i)
  i += 1
```

```python
# Print Number from 1 to N
n = int(input("Enter N: "))
i = 1
while i<=n:
  print(i)
  i += 1
```

```python
# Factorial of a number
n = int(input("Enter a Number: "))
fact = 1
while n > 0:
  fact *= n
  n -= 1
print("Factorial =",fact)
```

```python
# WAP for fibonacci series using While Loop
n = int(input("Enter number of terms: "))
a = 0
b = 1
i = 1
while i<=n:
  print(a, end=" ")
  c = a + b
  a = b
  b = c
  i += 1
```

```python
limit = int(input("Enter limit: "))
a = 0
b = 1
while a<=limit:
  print(a, end=" ")
  a, b=b, a+b
```

```python
# Reverse a Number
num = int(input("Enter a Number: "))
rev = 0
while num > 0:
  digit = num % 10
  rev = rev * 10 + digit
  num //= 10
print("Reversed Number =",rev)
```

```python
# Check Palindrome Number/String
num = int(input("Enter a Number: "))
temp = num
rev = 0
while num > 0:
  rev = rev * 10 + num % 10
  num //= 10
if temp == rev:
  print("Palindrome")
else:
  print("Not Palindrome")
```

```python
# Check Palindome Number
st = input("Enter the String:")
rev = st[::-1] # it helps to reverse the number
if st==rev:
  print("Palindrome")
else:
  print("Not Palindrome")
```

```python
s = "madam"
i, j = 0, len(s) - 1
is_palindrome = True
while i<j:
  if s[i] != s[j]:
    is_palindrome = False
    break
  i += 1
  j -= 1
if is_palindrome:
  print("Palindrome")
else:
  print("Not Palindrome")
```

```python
# Count Digits in a Number
num = int(input("Enter a Number:"))
count = 0
while num > 0:
  count += 1
  num = num // 10
print("Number of Digits=",count)
```

```python
# Search an element in a list
nums = [10, 20, 30, 40, 50]
key = int(input("Enter element to search:"))
i = 0
while i < len(nums):
  if nums[i] == key:
    print("Element found at index",i)
    break
  i += 1
else:
  print("Element not found")
```
